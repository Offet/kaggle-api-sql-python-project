```python
# pass an argument to read the csv file such that "Not available" and "unknown" are classified as Nan
df = pd.read_csv("orders.csv", na_values=["Not Available", "unknown"])
df["Ship Mode"].unique()
```

```python
# rename all columns to lowercase and replace " " with _
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.columns
```

```python
df.head(10)
```

```python
"""
It is seen in the df above that list price, and discount percentage are both listed, but the exact discount amount is unknown.
Find the discount amount and create a new column for it
"""

df["discount"] = df["list_price"]*df["discount_percent"]*.01
df
```

```python
"""
Sale Price: Taking a close look at the columns, there is a list price and a cost price, but there is no 
sales price available. Create a sales price column, and subtract the discount from the list
price to get how much items cost when they are on sales

Profits: There is no profits column available...create one 
"""

df["sale_price"] = df["list_price"] - df["discount"]

df["profit"] = df["sale_price"] - df["cost_price"]
df
```
